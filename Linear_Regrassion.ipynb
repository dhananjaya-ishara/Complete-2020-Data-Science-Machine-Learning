```python
import pandas
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from pandas import DataFrame
```

```python
data=pandas.read_csv("/content/sample_data/cost_revenue_clean.csv.csv")
```

```python
data
```

```python
data.describe()
```

```python
x=DataFrame(data,columns=['production_budget_usd'])
y=DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.title('production_budget_usd vs worldwide_gross_usd')
plt.xlabel('production_budget_usd $')
plt.ylabel('worldwide_gross_usd $')
plt.ylim(0,3000000000)
plt.xlim(0,400000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, regression.predict(x), color='red', linewidth=3)
plt.title('production_budget_usd vs worldwide_gross_usd')
plt.xlabel('production_budget_usd $')
plt.ylabel('worldwide_gross_usd $')
plt.ylim(0,3000000000)
plt.xlim(0,400000000)
plt.show()
```

```python
regression.score(x,y)
```
