#5. Understanding the Power Rule & Creating Charts with Subplots

##slop and derivatives
create a python funtion for the derivate f(x) called df(x)

```python
def df(x):
  return 2*x+1
```

##Calculus Power Rule

```python
#plot function and derivate side by side
plt.figure(figsize=[15,5])#with and hight size
plt.subplot(1,2,2)#(row,colum,possion)

#cost function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('f(x)=x^2+x+1')
plt.plot(x_1,f(x_1),color='blue',linewidth=3)
#derivate
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(-3,6)
plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()#like exel
plt.title('df(x)= 2*x+1')
plt.plot(x_1,df(x_1),color='red',linewidth=3)
plt.show()
```

Gradient descent

```python
new_x =3
privious_x=0
step_multiplier=0.1


for n in range(30):
  privious_x = new_x
  gradient = df(privious_x)
  new_x = privious_x - step_multiplier * gradient

print("the local minimum occurs at",new_x)
print("the local minimum occurs at",df(new_x))
print("the local minimum occurs at",f(new_x))
```

```python
new_x =3
privious_x=0
step_multiplier=0.1
precision=0.00001

for n in range(500):
  privious_x = new_x
  gradient = df(privious_x)
  new_x = privious_x - step_multiplier * gradient

  step_size=abs(new_x-privious_x)
  #print("step size",step_size)
  if step_size<precision:
    print("loop run this many times", n)
    break

print("the local minimum occurs at",new_x)
print("the local minimum occurs at",df(new_x))
print("the local minimum occurs at",f(new_x))
```

```python
new_x =3
privious_x=0
step_multiplier=0.1
precision=0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
  privious_x = new_x
  gradient = df(privious_x)
  new_x = privious_x - step_multiplier * gradient

  step_size=abs(new_x-privious_x)
  #print("step size",step_size)

  x_list.append(new_x)
  slope_list.append(df(new_x))

  if step_size<precision:
    print("loop run this many times", n)
    break

print("the local minimum occurs at",new_x)
print("the local minimum occurs at",df(new_x))
print("the local minimum occurs at",f(new_x))
```

```python
#superimpose the gradient descent calculationsplot plot
plt.figure(figsize=[15,5])#with and hight size
plt.subplot(1,2,1)#(row,colum,possion)

#cost function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('f(x)=x^2+x+1')
plt.plot(x_1,f(x_1),color='blue',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,f(values),color='red', s= 100, alpha=0.6)

#derivate
plt.subplot(1,2,2)
plt.xlim(-3,3)
plt.ylim(-3,6)
plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()#like exel
plt.title('df(x)= 2*x+1')
plt.plot(x_1,df(x_1),color='red',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,slope_list,color='blue', s= 100, alpha=0.6)

plt.show()
```

```python
#superimpose the gradient descent calculationsplot plot
plt.figure(figsize=[15,5])#with and hight size
plt.subplot(1,3,1)#(row,colum,possion)

#cost function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('f(x)=x^2+x+1')
plt.plot(x_1,f(x_1),color='blue',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,f(values),color='red', s= 100, alpha=0.6)

#derivate
plt.subplot(1,3,2)
plt.xlim(-3,3)
plt.ylim(-3,6)
plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()#like exel
plt.title('df(x)= 2*x+1')
plt.plot(x_1,df(x_1),color='red',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,slope_list,color='blue', s= 100, alpha=0.6)


#derivative (close up)
plt.subplot(1,3,3)
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.grid()#like exel
plt.title("gradient descent coose up")

plt.plot(x_1,df(x_1),color='skyblue',linewidth=6)
plt.scatter(x_list,slope_list,color='red', s= 300, alpha=0.6)

plt.show()
```

```python
#Gradient Descent
new_x =-3#chane in 3 to -3
privious_x=0
step_multiplier=0.1
precision=0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
  privious_x = new_x
  gradient = df(privious_x)
  new_x = privious_x - step_multiplier * gradient

  step_size=abs(new_x-privious_x)
  #print("step size",step_size)

  x_list.append(new_x)
  slope_list.append(df(new_x))

  if step_size<precision:
    print("loop run this many times", n)
    break

print("the local minimum occurs at",new_x)
print("the local minimum occurs at",df(new_x))
print("the local minimum occurs at",f(new_x))
```

```python
#superimpose the gradient descent calculationsplot plot
plt.figure(figsize=[15,5])#with and hight size
plt.subplot(1,3,1)#(row,colum,possion)

#cost function
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('f(x)=x^2+x+1')
plt.plot(x_1,f(x_1),color='blue',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,f(values),color='red', s= 100, alpha=0.6)

#derivate
plt.subplot(1,3,2)
plt.xlim(-3,3)
plt.ylim(-3,6)
plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()#like exel
plt.title('df(x)= 2*x+1')
plt.plot(x_1,df(x_1),color='red',linewidth=3)

values = np.array(x_list)
plt.scatter(x_list,slope_list,color='blue', s= 100, alpha=0.6)


#derivative (close up)
plt.subplot(1,3,3)
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.grid()#like exel
plt.title("gradient descent coose up")

plt.plot(x_1,df(x_1),color='skyblue',linewidth=6)
plt.scatter(x_list,slope_list,color='red', s= 300, alpha=0.6)

plt.show()
```
